#### Taking Input to the Linked List (O(n))

```python
def printLL(pair):
    curr = pair.head
    while curr is not None:
        print(str(curr.data) + "-->",end = " ")
        curr = curr.next
    print("None")    
    return
        
def takeInput():
    pair = htPair()
    pair.head = None
    pair.tail = None
    inputList = list(map(int,input().split()))
    for currData in inputList:
        if currData == -1:
            break
        newNode = Node(currData) 
        if pair.head is None:
            pair.head = newNode
            pair.tail = newNode
        else:
            pair.tail.next = newNode
            pair.tail = newNode
            
            
    return pair 

pair = takeInput()
printLL(pair)
print(id(takeInput))
```

#### Inserting a Node at ith Position.

```python
def insertIth(pair,i,data):
    newNode = Node(data)
    if i == 0:
        newNode.next = pair.head
        pair.head = newNode
        return
    
    curr = pair.head
    for j in range(i):
        prev = curr
        curr = curr.next
        
    newNode.next = curr
    prev.next = newNode
    return
    
```

#### Delete ith Index Node

```python
def deleteIth(pair,i):
    ans = None
    if i == 0:
        ans = pair.head.data
        temp = pair.head
        pair.head = pair.head.next
        del temp
        return ans
    
    prev = None
    curr = pair.head
    for j in range(i):
        prev = curr
        curr = curr.next
    
    ans = curr.data
    if curr.next == None:
        prev.next = None
        pair.tail = prev
    else:
        prev.next = curr.next
        
    del curr    
    return ans    
```

#### Reverse Linked List

```python
from time import sleep
def ReverseLL(head):
    if head is None:
        return head
    if head.next is None:
        return head
    
    curr = head.next
    node = Node(curr.data)
    node.next = curr.next
    curr.next = head
    head.next = None
    
    while node.next is not None:
        temp = node.next
        temp1 = node.next.next
        node.next.next = curr
        curr = temp
        
        node.next = temp1
        print(curr.data)
        
        sleep(1)
    print("Exiting while loop")
    return curr
        
pair2 = takeInput()
printLL(pair2)
pair2.head = ReverseLL(pair2.head)
printLL(pair2)
```

#### find Middle term of Linked List

```python
def getMid(head):
    if head is None:
        return head
    if head.next is None:
        return head
    p = head
    q = head
    while True:
        if q.next is None:
            break
        if q.next.next is None:
            break
            
        p = p.next
        q = q.next.next
        
    return p


printLL(pair3)
ans = getMid(pair3.head)
print(ans.data)
```

#### Merge two Sorted Linked List

```python
def mergeLL(head1,head2):
    if head1 is None:
        return head2
    if head2 is None:
        return head1
    
    head = None
    if head1.data < head2.data:
        head = head1
        head1 = head1.next
    else:
        head = head2
        head2 = head2.next
    tail = head
    while (head1 is not None) and (head2 is not None):
        if head1.data < head2.data:
            temp = head1
            head1 = head1.next
            tail.next = temp
            tail = tail.next
        else:
            temp = head2 
            head2 = head2.next
            tail.next = temp
            tail = tail.next
            
    tail.next = None                
    if head1 is not None:
        tail.next = head1
        
    if  head2 is not None:
        tail.next = head2
        
    return head
             
```

```python
pair1 = takeInput()
pair2 = takeInput()

ans = mergeLL(pair1.head,pair2.head)
print("linked list 1:")
printLL(pair1)
print("Linked list 2:")
printLL(pair2)
pair1.head = mergeLL(pair1.head,pair2.head)
print("merged Linked list: ")
printLL(pair1)
```
