#### Implementation of Queue through List

```python
class Queue:
    def __init__(self):
        self.__arr = []
        self.__front = 0
        self.__count = 0
        
    def push(self,item):
        self.__arr.append(item)
        self.__count += 1
        
    def dequeue(self):
        if self.__count == 0:
            return -1
        ans = self.__arr[self.__front]
        self.__front += 1
        self.__count -= 1
        return ans
    
    def front(self):
        if self.__count == 0:
            return -1
        return self.__arr[self.__front]
    
    def size(self):
        return self.__count
    
    def isEmpty(self):
        return self.size() == 0
    
    def printQ(self):
        i = self.__front
        for j in range(self.__count):
            print(self.__arr[i])
            i += 1
    
q = Queue()
q.push(1)
q.push(2)
q.push(3)
q.push(4)
q.push(5)
print("elements in queue:")
q.printQ()

print("size of queue:")
print(q.size())
print()
print(q.front())
print()

print(q.dequeue())
print()
while not q.isEmpty():
    print(q.dequeue())
    
    
```

#### Implementation of Queue through Linked List

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        self.__count = 0
        
class Queue:
    def __init__(self):
        self.__head = None
        self.__count = 0
        self.__tail = None
        
    def push(self,item):
        if self.__head is None:
            self.__head = Node(item)
            self.__tail = self.__head
            self.__count += 1
            return
        newNode = Node(item)
        self.__tail.next = newNode
        self.__tail = newNode
        self.__count += 1
        return
    
    def enQ(self):
        if self.__head is None:
            return "Queue is empty"
        ans = self.__head.data
        temp = self.__head
        self.__head = self.__head.next
        del temp
        return ans
    
    def front(self):
        if self.__head is None:
            return "Queue is empty"
        return self.__head.data
    
    def size(self):
        return self.__count
    
    def isEmpty(self):
        return self.__count == 0
    
    def printQ(self):
        if self.__head is None:
            print("Queue is empty")
            return
        curr = self.__head
        while curr is not None:
            print(curr.data)
            curr = curr.next

q = Queue()
print("Elements in the Queue:")
q.printQ()
print()
print("Removing element from Queue:")
print(q.enQ())
print()
print("Getting the front element of the Queue:")
print(q.front())
print()
q.push(1)
q.push(2)
q.push(3)
q.push(4)
q.push(5)
print()
print("After pushing elements in the Queue")
q.printQ()
print()
print("Getting the Front element of the Queue:")
print(q.front())
print()
print("removing front element from the Queue:")
print(q.enQ())
print()

print("Elements in the Queue:")
q.printQ()

q.push(1)
q.push(6)
print("after pushing Elements in the Queue:")
q.printQ()

            
    
    
        
        
        
```
